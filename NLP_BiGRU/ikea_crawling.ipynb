```
import codecs
from bs4 import BeautifulSoup
from urllib.request import urlopen
from google.colab import drive

drive.mount('/content/gdrive')

html_gagus = urlopen('https://www.ikea.com/kr/ko/cat/gagu-fu001/') 
soup_gagus = BeautifulSoup(html_gagus, 'html.parser')
gagus = soup_gagus.find_all(class_='vn-link vn__nav__link vn-6-grid-gap')
furniture_category = ['가죽소파', '소파베드', '더블/퀸/킹침대', '싱글/수퍼싱글+침대', '디반침대', 
                      '옷장', '어린이옷장', '바테이블', '커피테이블/보조테이블', '식탁',  
                      '식탁의자', '바의자', '사무용의자', '스툴/벤치', '어린이의자', '영아용의자', 
                      '가죽암체어', '라탄암체어', '리클라이너', '커튼', '샤워커튼', '블라인드'] # 22개

temp = []
f_explanes = []
for i in range(0,len(furniture_category)):
    f_explanes.append([])

for i in range(0,11):
    if i != 10:
        link_in_gagu = gagus[i].attrs['href']
        html_in_gagu = urlopen(link_in_gagu)
        soup_in_gagu = BeautifulSoup(html_in_gagu, 'html.parser')
        in_in_gagus = soup_in_gagu.find_all(class_='vn-link vn__nav__link vn-6-grid-gap')
    else:
        link_in_gagu = "https://www.ikea.com/kr/ko/cat/keoteun-beulraindeu-tl002/" 
        html_in_gagu = urlopen(link_in_gagu)
        soup_in_gagu = BeautifulSoup(html_in_gagu, 'html.parser')
        in_in_gagus = soup_in_gagu.find_all(class_='vn-link vn__nav__link vn-3-grid-gap')
    
    for in_in_gagu in in_in_gagus:
        try:
            link_in_in_gagu = in_in_gagu.attrs['href']
            html_in_in_gagu = urlopen(link_in_in_gagu)
            soup_in_in_gagu = BeautifulSoup(html_in_in_gagu, 'html.parser')
            products = soup_in_in_gagu.find_all(class_='plp-fragment-wrapper')

            for product in products:
                a = product.find("a")
                link_product = a.attrs['href']
                html_product = urlopen(link_product)
                soup_product = BeautifulSoup(html_product, 'html.parser')
                name_temp = soup_product.find_all(class_='bc-breadcrumb__list-item')
                name = name_temp[3].get_text().rstrip('\n')
                name = name.lstrip('\n')
                name = name.replace(" ","")      
                if name == '패브릭카우치':
                    name = '패브릭소파'
                elif name == '천연/인조가죽소파' or name == '천연가죽카우치':
                    name = '가죽소파'               
                elif name == '어린이암체어':
                    name = '어린이의자'            
                elif name == '서랍장':
                    name = '수납장'
                if name not in furniture_category:
                    continue
                temp.append(name)
                explanes = soup_product.find_all(class_='range-revamp-product-details__paragraph')   
                for i in range(0,len(explanes)):
                    explane = explanes[i].get_text()
                    if explane not in f_explanes[furniture_category.index(name)]:
                        f_explanes[furniture_category.index(name)].append(explane)
                    
        except:
            continue
```
