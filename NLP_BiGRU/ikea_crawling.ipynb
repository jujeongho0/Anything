```
import codecs
from bs4 import BeautifulSoup
from urllib.request import urlopen
from google.colab import drive

drive.mount('/content/gdrive')
f_name = codecs.open("/content/gdrive/My Drive/crawling_ikea/NAME.txt", 'w', 'utf-8')
f_explane = codecs.open("/content/gdrive/My Drive/crawling-ikea/EXPLANE.txt", 'w', 'utf-8')
f_plus = codecs.open("/content/gdrive/My Drive/crawling-ikea/PLUS
.txt", 'w', 'utf-8')

response = urlopen('https://www.ikea.com/kr/ko/cat/jepum-products/')
soup = BeautifulSoup(response, 'html.parser')
categorys = soup.find_all(class_='vn-link vn-nav__link')

for category in categorys:
    link = category.attrs['href']
    name = category.get_text()
    response_ = urlopen(link)
    soup_ = BeautifulSoup(response_, 'html.parser')       
    products = soup_.find_all(class_='plp-product-list__products')
    
    for product in products:
        a = product.find_all("a")
        
        for a_ in a:
            link_ = a_.attrs['href']
            response__ = urlopen(link_)
            soup__ = BeautifulSoup(response__, 'html.parser')
            name_ = soup__.find("title").get_text() 
            explanes = soup__.find_all(class_='range-revamp-product-details__paragraph')          
            for explane in explanes:
                explane_ = explane.get_text()
                print(name_ + "은/는 " + explane_)
                try:
                    f_name.write(name_ + "\n") 
                    f_explane.write(explane_ + "\n") 
                    f_plus.write(name_ + "은/는 " + explane_ + "\n") 
                except:
                    continue

f_name.close()
f_explane.close()
f_plus.close()
```
