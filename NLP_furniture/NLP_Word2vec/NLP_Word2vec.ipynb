```
!pip install konlpy
import codecs
from gensim.models import Word2Vec
from konlpy.tag import okt
from google.colab import drive
import gensim
drive.mount('/content/gdrive')
okt = Okt()

furniture_category = ['패브릭소파', '가죽소파', '풋스툴/쿠션형스툴', '소파베드',
                      '더블/퀸/킹침대', '싱글/수퍼싱글+침대', '유아용침대', '로프트/2층침대', 
                      '디반침대', '어린이침대', '옷장', '어린이옷장', '바테이블', 
                      '커피테이블/보조테이블', '식탁', '책상/컴퓨터책상', '화장대', 
                      '어린이테이블', '기저귀교환대', '식탁의자', '바의자', '카페의자', 
                      '사무용의자', '스툴/벤치', '어린이의자', '영아용의자', '패브릭암체어', 
                      '가죽암체어', '라탄암체어', '리클라이너', '수납가구', '커튼', 
                      '샤워커튼', '블라인드'] # 34개

fread = codecs.open("/content/gdrive/My Drive/word2vec/PLUS.txt", 'r', 'utf-8')
result = []

while True:
    try:
        line = fread.readline()
        if not line: break
        sentence = okt.pos(line)
        temp = []
        for morpheme in sentence:
                if 'Noun' in morpheme[1]:
                    temp.append(morpheme[0])                
                elif 'Verb' in morpheme[1]:
                    temp.append(morpheme[0])                
                elif 'Adjective' in morpheme[1]:
                    temp.append(morpheme[0])                
        if temp:
            result.append(temp)
    except:
        continue

fread.close()
model = Word2Vec(result, size=100, window=10, min_count=10, workers=4, sg=1)
model.save('/content/gdrive/My Drive/word2vec/word2vec')
```
